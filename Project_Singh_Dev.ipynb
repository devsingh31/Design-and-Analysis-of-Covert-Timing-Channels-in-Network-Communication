```r
  buffer_behavior <- function(B, lambda, i, m) {
  inter_packet_delays <- rexp(m, rate = lambda)
  buffer <- numeric(i)

  overflow <- 0
  underflow <- 0

  for (j in 1:m) {
    if (length(buffer) >= B) {
      overflow <- overflow + 1
    }

    if (length(buffer) == 0) {
      underflow <- underflow + 1
    }

    buffer <- c(buffer, inter_packet_delays[j])


  }

  probability_overflow <- overflow / m
  probability_underflow <- underflow / m

  return(c(probability_overflow, probability_underflow))
}

B <- 20
lambda <- 1
message_sizes <- c(16, 32)
buffer_sizes <- c(2, 6, 10, 14, 18)

for (i in buffer_sizes) {
  for (m in message_sizes) {
    probabilities <- buffer_behavior(B, lambda, i, m)
    cat("i =", i, ", m =", m)
    cat("Overflow:", probabilities[1])
    cat("Underflow:", probabilities[2])
    cat("\n")
  }
}
```

```r
buffer_uniform <- function(B, i, m) {
  inter_packet_delays <- runif(m, min = 0, max = 1)

  buffer <- numeric(i)

  overflow <- 0
  underflow <- 0

  for (j in 1:m) {
    if (length(buffer) >= B) {
      overflow <- overflow + 1
    }

    if (length(buffer) == 0) {
      underflow <- underflow + 1
    }

    buffer <- c(buffer, inter_packet_delays[j])

    buffer <- tail(buffer, B)
  }

  probability_overflow <- overflow / m
  probability_underflow <- underflow / m

  return(c(probability_overflow, probability_underflow))
}

B <- 20
message_sizes <- c(16, 32)
buffer_sizes <- c(2, 6, 10, 14, 18)

for (i in buffer_sizes) {
  for (m in message_sizes) {
    probabilities <- buffer_uniform(B, i, m)
    cat("For B =", B, ", Uniform distribution, i =", i, ", m =", m)
    cat("Overflow:", probabilities[1])
    cat("Underflow:", probabilities[2])
    cat("\n")
  }
}
```
